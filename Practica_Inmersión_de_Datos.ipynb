```python
import pandas as pd
nivel_estudios = pd.read_csv('/content/drive/MyDrive/Cursos/inmersion-datos/Censo2017/distrito_nivel_estudios.csv',sep=',',encoding='latin-1')
nivel_estudios.sample(10)
```

```python
nivel_estudios.dropna(subset=['Sin Nivel','Inicial','Primaria','Secundaria','Basica especial','Superior no universitaria incompleta','Superior no universitaria completa','Superior universitaria incompleta','Superior universitaria completa','Maestria / Doctorado'])

```

```python
nivel_estudios['Sin Nivel'] = nivel_estudios['Sin Nivel'].str.replace('.','', regex = True)
nivel_estudios['Inicial'] = nivel_estudios['Inicial'].str.replace('.','', regex = True)
nivel_estudios['Primaria'] = nivel_estudios['Primaria'].str.replace('.','', regex = True)
nivel_estudios['Secundaria'] = nivel_estudios['Secundaria'].str.replace('.','', regex = True)
nivel_estudios['Basica especial'] = nivel_estudios['Basica especial'].str.replace('.','', regex = True)
nivel_estudios['Superior no universitaria incompleta'] = nivel_estudios['Superior no universitaria incompleta'].str.replace('.','', regex = True)
nivel_estudios['Superior no universitaria completa'] = nivel_estudios['Superior no universitaria completa'].str.replace('.','', regex = True)
nivel_estudios['Superior universitaria incompleta'] = nivel_estudios['Superior universitaria incompleta'].str.replace('.','', regex = True)
nivel_estudios['Superior universitaria completa'] = nivel_estudios['Superior universitaria completa'].str.replace('.','', regex = True)
nivel_estudios['Maestria / Doctorado'] = nivel_estudios['Maestria / Doctorado'].str.replace('.','', regex = True)

nivel_estudios['Sin Nivel'] = nivel_estudios['Sin Nivel'].str.replace(' ','', regex = True)
nivel_estudios['Inicial'] = nivel_estudios['Inicial'].str.replace(' ','', regex = True)
nivel_estudios['Primaria'] = nivel_estudios['Primaria'].str.replace(' ','', regex = True)
nivel_estudios['Secundaria'] = nivel_estudios['Secundaria'].str.replace(' ','', regex = True)
nivel_estudios['Basica especial'] = nivel_estudios['Basica especial'].str.replace(' ','', regex = True)
nivel_estudios['Superior no universitaria incompleta'] = nivel_estudios['Superior no universitaria incompleta'].str.replace(' ','', regex = True)
nivel_estudios['Superior no universitaria completa'] = nivel_estudios['Superior no universitaria completa'].str.replace(' ','', regex = True)
nivel_estudios['Superior universitaria incompleta'] = nivel_estudios['Superior universitaria incompleta'].str.replace(' ','', regex = True)
nivel_estudios['Superior universitaria completa'] = nivel_estudios['Superior universitaria completa'].str.replace(' ','', regex = True)
nivel_estudios['Maestria / Doctorado'] = nivel_estudios['Maestria / Doctorado'].str.replace(' ','', regex = True)

nivel_estudios['Sin Nivel'] = nivel_estudios['Sin Nivel'].str.replace('-','', regex = True)
nivel_estudios['Inicial'] = nivel_estudios['Inicial'].str.replace('-','', regex = True)
nivel_estudios['Primaria'] = nivel_estudios['Primaria'].str.replace('-','', regex = True)
nivel_estudios['Secundaria'] = nivel_estudios['Secundaria'].str.replace('-','', regex = True)
nivel_estudios['Basica especial'] = nivel_estudios['Basica especial'].str.replace('-','', regex = True)
nivel_estudios['Superior no universitaria incompleta'] = nivel_estudios['Superior no universitaria incompleta'].str.replace('-','', regex = True)
nivel_estudios['Superior no universitaria completa'] = nivel_estudios['Superior no universitaria completa'].str.replace('-','', regex = True)
nivel_estudios['Superior universitaria incompleta'] = nivel_estudios['Superior universitaria incompleta'].str.replace('-','', regex = True)
nivel_estudios['Superior universitaria completa'] = nivel_estudios['Superior universitaria completa'].str.replace('-','', regex = True)
nivel_estudios['Maestria / Doctorado'] = nivel_estudios['Maestria / Doctorado'].str.replace('-','', regex = True)
```

```python
ubicacion = nivel_estudios.Ubicacion.str.split(',',expand=True)
nivel_estudios['Dpto']=ubicacion[0]

nivel_estudios['Secundaria'] = pd.to_numeric(nivel_estudios['Secundaria'], errors='coerce')
nivel_estudios['Primaria'] = pd.to_numeric(nivel_estudios['Primaria'], errors='coerce')
nivel_estudios['Sin Nivel'] = pd.to_numeric(nivel_estudios['Sin Nivel'], errors='coerce')
nivel_estudios['Basica especial'] = pd.to_numeric(nivel_estudios['Basica especial'], errors='coerce')
nivel_estudios['Secundaria'] = nivel_estudios['Secundaria'] + nivel_estudios['Primaria'] + nivel_estudios['Sin Nivel'] + nivel_estudios['Basica especial']

nivel_estudios['Superior no universitaria incompleta'] = pd.to_numeric(nivel_estudios['Superior no universitaria incompleta'], errors='coerce')
nivel_estudios['Superior no universitaria completa'] = pd.to_numeric(nivel_estudios['Superior no universitaria completa'], errors='coerce')
nivel_estudios['Superior universitaria incompleta'] = pd.to_numeric(nivel_estudios['Superior universitaria incompleta'], errors='coerce')
nivel_estudios['Superior universitaria completa'] = pd.to_numeric(nivel_estudios['Superior universitaria completa'], errors='coerce')
nivel_estudios['Maestria / Doctorado'] = pd.to_numeric(nivel_estudios['Maestria / Doctorado'], errors='coerce')
nivel_estudios['Maestria / Doctorado'] = nivel_estudios['Maestria / Doctorado'] + nivel_estudios['Superior universitaria completa'] + nivel_estudios['Superior universitaria incompleta']+ nivel_estudios['Superior no universitaria completa'] + nivel_estudios['Superior no universitaria incompleta']


df2 = nivel_estudios.groupby('Dpto')['Secundaria', 'Maestria / Doctorado'].sum()
df2 = df2.reset_index()
df2 = df2.rename(columns={'Secundaria': 'Sin_Estudios_Superiores', 'Maestria / Doctorado': 'Con_Estudios_Superiores'})
df2.head(26)
```

```python
# total
df2 = df2.assign(Total=df2.Sin_Estudios_Superiores + df2.Con_Estudios_Superiores)

# porcentaje de Sin_Estudios_Superiores
df2 = df2.assign(Porcentaje_Sin_Estudios_Superiores=df2.Sin_Estudios_Superiores.div(df2.Total)*100)

# porcentaje de Con_Estudios_Superiores
df2 = df2.assign(Porcentaje_Con_Estudios_Superiores=df2.Con_Estudios_Superiores.div(df2.Total)*100)

df2.head(10)
```

```python
df2.info()
```

```python
import matplotlib.pyplot as plt
```

```python
df_sorted = df2.sort_values('Porcentaje_Con_Estudios_Superiores', ascending=False)
fig, ax = plt.subplots()
bar_width = 0.8 
plt.bar(df_sorted ['Dpto'], df_sorted ['Porcentaje_Sin_Estudios_Superiores'], width=bar_width, label='Sin estudios superiores')
plt.bar(df_sorted ['Dpto'], df_sorted ['Porcentaje_Con_Estudios_Superiores'], bottom=df_sorted ['Porcentaje_Sin_Estudios_Superiores'], width=bar_width, label='Con estudios superiores')
plt.xlabel('Departamento')
plt.ylabel('Porcentaje')
ax.legend(fontsize=7)
plt.legend()
#import numpy as np
plt.xticks( rotation=75)
fig.set_size_inches(14, 8) 
plt.show()
```
