```python
import pandas as pd
nivel_estudios = pd.read_csv('/content/drive/MyDrive/Cursos/inmersion-datos/Censo2017/distrito_nivel_estudios.csv',sep=',',encoding='latin-1')
nivel_estudios.columns
```

```python
nivel_estudios.head()
```

```python
columnas = {'Unnamed: 0':'Ubicacion',
            'Sin Nivel':'Sin estudios formales', 
            'BÃ¡sica especia': 'Basica Especial',
            'Superior no universitaria incompleta': 'No Universitaria Incompleta',
            'Superior no universitaria completa':'No Universitaria Completa',
            'Superior universitaria incompleta': 'Universitaria Incompleta',
            'Superior universitaria completa': 'Universitaria Completa',
            'MaestrÃ­a / Doctorad':'Maestría Doctorado'}
nivel_estudios = nivel_estudios.rename(columns=columnas)
nivel_estudios.head(10)
```

```python
valor = nivel_estudios.Ubicacion.str.split(', ',expand=True)
nivel_estudios['Departamento']=valor[0]
nivel_estudios.sample(10)
```

```python
nivel_estudios['Inicial'] = nivel_estudios['Inicial'].str.replace(' ', '').str.replace('-', '')
nivel_estudios['Inicial'] = pd.to_numeric(nivel_estudios['Inicial'], errors='coerce')
nivel_estudios = nivel_estudios.dropna(subset=['Inicial']).astype({'Inicial':int})
j = nivel_estudios.groupby('Departamento').agg({'Inicial': 'sum'})
j.iloc[0:20]



```
